```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Les fonctions Seaborn ont presque toutes la même structure : sns.fonction(x=..., y=..., data=..., [hue=..., size=..., style=...])

# 1. Vue d'ensemble d'un jeux de données  <a class="anchor" id="1"></a>
* **sns.pairplot**(dataset) : permet de représenter toutes les relations entre les variables d'un jeux de données

```python
# Chargement d'un jeux de données
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

# 2. Visualiser des catégories  <a class="anchor" id="1"></a>
* **sns.catplot**() : nuage de point par catégories
* **sns.boxplot**() : box plot
* **sns.violinplot**() : violon plot
* **sns.barplot**() : bar plot
* **sns.swarmplot**() : swarn plot

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.catplot(x="survived", y="age", data=titanic, hue="sex")
plt.show()
```

```python
plt.figure(figsize=(30, 8))
sns.boxplot(x="age", y="fare", data=titanic)
plt.show()
```

```python
sns.swarmplot(x="sex", y="age", hue='survived', data=titanic)
plt.show()
```

# 3. Visualiser des distributions  <a class="anchor" id="3"></a>
* **sns.distplot**() : représente la distribution d'une variable.
* **sns.joinplot**() : distribution selon 2 variables.
* **sns.heatmap**() : heatmap
* **sns.clustermap**() : heatmap clusteurisé

```python
sns.distplot(titanic['fare'])
plt.show()
```

```python
sns.jointplot(x='age', y='fare', data=titanic, kind='hex')
plt.show()
```

```python
# Affiche la matrice de corrélation
sns.heatmap(titanic.corr())
plt.show()
```

```python
# Affiche la matrice de corrélation cluteurisé
plt.figure(figsize=(5, 5))
sns.clustermap(titanic.corr())
plt.show()
```
